```python
#Evaluation Metric MAPE = Mean absolute percentage error
def MAPE(pred,org):
    temp = np.abs((pred-org)/org)*100
    return np.mean(temp)
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing # SES
ses_model = SimpleExpSmoothing(df_train["Sales"]).fit(smoothing_level=0.2)
pred_ses = ses_model.predict(start = df_test.index[0],end = df_test.index[-1])
MAPE(pred_ses,df_test.Sales)
```

```python
# Holts Exponential Smoothing method 
from statsmodels.tsa.holtwinters import Holt # Holts Exponential Smoothing
hw_model = Holt(df_train["Sales"]).fit(smoothing_level=0.1, smoothing_trend=0.2)
pred_hw = hw_model.predict(start = df_test.index[0],end = df_test.index[-1])
MAPE(pred_hw,df_test.Sales)
```

```python
#Holts winter exponential smoothing with additive seasonality and additive trend
from statsmodels.tsa.holtwinters import ExponentialSmoothing
hwe_model_add_add = ExponentialSmoothing(df_train["Sales"],seasonal="add",trend="add",seasonal_periods=12).fit(smoothing_level=0.1, smoothing_trend=0.2) #add the trend to the model
pred_hwe_add_add = hwe_model_add_add.predict(start = df_test.index[0],end = df_test.index[-1])
MAPE(pred_hwe_add_add,df_test.Sales)
```

```python
# Holts winter exponential smoothing with multiplicative seasonality and additive trend
hwe_model_mul_add = ExponentialSmoothing(df_train["Sales"],seasonal="mul",trend="add",seasonal_periods=12).fit(smoothing_level=0.1, smoothing_trend=0.2) 
pred_hwe_mul_add = hwe_model_mul_add.predict(start = df_test.index[0],end = df_test.index[-1])
MAPE(pred_hwe_mul_add,df_test.Sales)
```

```python
#RMSE
from sklearn.metrics import mean_squared_error
rmse_hwe_mul_add = sqrt(mean_squared_error(pred_hwe_mul_add,df_test.Sales))
rmse_hwe_mul_add
```

```python
#Splitting data'
coke_train = coke.head(110) # training data
coke_test = coke.tail(14) # test Data
```

```python
#Linear Model
import statsmodels.formula.api as smf 

linear_model = smf.ols('Sales~t',data=coke_train).fit()
pred_linear =  pd.Series(linear_model.predict(pd.DataFrame(coke_test['t'])))
rmse_linear = np.sqrt(np.mean((np.array(coke_test['Sales'])-np.array(pred_linear))**2))
rmse_linear
```

```python
#Exponential
Exp = smf.ols('log_sales~t',data=coke_train).fit()
pred_Exp = pd.Series(Exp.predict(pd.DataFrame(coke_test['t'])))
rmse_Exp = np.sqrt(np.mean((np.array(coke_test['Sales'])-np.array(np.exp(pred_Exp)))**2))
rmse_Exp
```

```python
#Quadratic 
Quad = smf.ols('Sales~t+t_sq',data=coke_train).fit()
pred_Quad = pd.Series(Quad.predict(coke_test[["t","t_sq"]]))
rmse_Quad = np.sqrt(np.mean((np.array(coke_test['Sales'])-np.array(pred_Quad))**2))
rmse_Quad
```

```python
#Additive seasonality 

add_sea = smf.ols('Sales~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=coke_train).fit()
pred_add_sea = pd.Series(add_sea.predict(coke_test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov']]))
rmse_add_sea = np.sqrt(np.mean((np.array(coke_test['Sales'])-np.array(pred_add_sea))**2))
rmse_add_sea
```

```python
#Additive Seasonality Quadratic 

add_sea_Quad = smf.ols('Sales~t+t_sq+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data=coke_train).fit()
pred_add_sea_quad = pd.Series(add_sea_Quad.predict(coke_test[['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','t','t_sq']]))
rmse_add_sea_quad = np.sqrt(np.mean((np.array(coke_test['Sales'])-np.array(pred_add_sea_quad))**2))
rmse_add_sea_quad
```

```python
##Multiplicative Seasonality

Mul_sea = smf.ols('log_sales~Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data = coke_train).fit()
pred_Mult_sea = pd.Series(Mul_sea.predict(coke_test))
rmse_Mult_sea = np.sqrt(np.mean((np.array(coke_test['Sales'])-np.array(np.exp(pred_Mult_sea)))**2))
rmse_Mult_sea
```

```python
#Multiplicative Additive Seasonality 

Mul_Add_sea = smf.ols('log_sales~t+Jan+Feb+Mar+Apr+May+Jun+Jul+Aug+Sep+Oct+Nov',data = coke_train).fit()
pred_Mult_add_sea = pd.Series(Mul_Add_sea.predict(coke_test))
rmse_Mult_add_sea = np.sqrt(np.mean((np.array(coke_test['Sales'])-np.array(np.exp(pred_Mult_add_sea)))**2))
rmse_Mult_add_sea
```

```python
#Compareing the results 
data = {"MODEL":pd.Series(["rmse_linear","rmse_Exp","rmse_Quad","rmse_add_sea","rmse_add_sea_quad","rmse_Mult_sea","rmse_Mult_add_sea"]),"RMSE_Values":pd.Series([rmse_linear,rmse_Exp,rmse_Quad,rmse_add_sea,rmse_add_sea_quad,rmse_Mult_sea,rmse_Mult_add_sea])}
table_rmse=pd.DataFrame(data)
table_rmse.sort_values(['RMSE_Values'])
```
