##Perform Clustering(Hierarchical, Kmeans & DBSCAN) 
for the crime data and identify the number of clusters formed and draw inferences.

Data Description:
Murder -- Muder rates in different places of United States
Assualt- Assualt rate in different places of United States
UrbanPop - urban population in different places of United States
Rape - Rape rate in different places of United States


```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

##-----Exploratory Data Analysis-------

```python
#normalizing data
```

```python
from sklearn.preprocessing import StandardScaler
norm_cd=StandardScaler().fit_transform(cd)
norm_cd
```

```python
# clustering the normalised data with dbscan
```

```python
from sklearn.cluster import DBSCAN
dbs=DBSCAN(eps=1,min_samples=4)
dbs.fit(norm_cd)
```

```python
dbs.labels_
```

```python
#adding array of cluster values to the dataset
cd['clusters']=dbs.labels_
cd
```

```python
#grouping by clusters
cd.groupby('clusters').agg(['mean']).reset_index()
```

```python
# Plotting the Clusters
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))  
plt.scatter(cd['clusters'],cd['UrbanPop'], c=dbs.labels_) 
```

```python
#HEIRARCHAL CLUSTER and creating dendrogram
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(norm_cd, method='single'))
```

```python
#CLUSTERING WITH KMEANS
from sklearn.cluster import KMeans
kmeans=KMeans().fit(norm_cd)
score=[]
K=range(1,20)

for i in K:
    kmeans=KMeans(n_clusters=i,init="k-means++",random_state=33,n_init='auto')
    kmeans.fit(norm_cd)
    score.append(kmeans.inertia_)
```

```python
#visualize using the Elbow Graph to find optimum number of  clusters (K value) from K values rang;
plt.plot(K,score,color="red")
plt.xlabel("k value")
plt.ylabel("wcss value") # within-cluster sum-of-squares criterion 
plt.show()
```

```python
 #Fitting K-Means to the dataset with 4 clusters
kmeans = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42)
y_kmeans = kmeans.fit_predict(norm_cd)
y_kmeans
```

```python
# New list called cluster
cluster = list(y_kmeans)
# Adding cluster to our data set
cd['cluster'] = cluster
cd
```

```python
#grouping by clusters
cd.groupby('cluster').agg(['mean']).reset_index()
```

```python
import seaborn as sns

plt.figure(figsize=(12,6))
sns.scatterplot(x=cd['Murder'], y = cd['Assault'],hue=y_kmeans)
```

```python

plt.figure(figsize=(10, 7))  
plt.scatter(cd['Murder'],cd['UrbanPop'], c=cluster) 
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(cd['Murder'],cd['Rape'], c=cluster) 
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(cd['Assault'],cd['Rape'], c=cluster) 
```
