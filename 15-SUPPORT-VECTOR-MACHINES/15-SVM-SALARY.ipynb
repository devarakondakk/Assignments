```python
##Preprocessing the data. As, there are categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
col = ["workclass","education","maritalstatus","occupation","relationship","race","sex","native"]
for i in col:
        train[i]= le.fit_transform(train[i])
        test[i]=le.fit_transform(test[i])
```

```python
colum = train.columns
colum
```

```python
# splitting test and train
x_train = train[colum[0:13]]
y_train = train[colum[13]]
x_test = test[colum[0:13]]
y_test = test[colum[13]]
```

```python
##Normalmization
def nor(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
x_train = nor(x_train)
x_test =  nor(x_test)
```

```python
x_train.shape, y_train.shape, x_test.shape, y_test.shape
```

```python
from sklearn.svm import SVC
clf = SVC(kernel='linear', gamma=3)
clf.fit(x_train,y_train)
```

```python
y_pred_train = clf.predict(x_train)
y_pred_test  = clf.predict(x_test)
```

```python
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, y_pred_test)
print(cm)
```

```python
print("Training Accuracy :",metrics.accuracy_score(y_train, y_pred_train).round(2))
print("Testing Accuracy:",metrics.accuracy_score(y_test, y_pred_test).round(2))
```
