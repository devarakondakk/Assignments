```python
import numpy as np
import statsmodels.api as stm
import matplotlib.pyplot as plt
stm.qqplot(np.log(df['YearsExperience']), line = 'r')
plt.title('No transformation')
stm.qqplot(np.sqrt(df['YearsExperience']), line = 'r')
plt.title('Log transformation')
stm.qqplot(np.sqrt(df['YearsExperience']), line = 'r')
plt.title('Square root transformation')
stm.qqplot(np.cbrt(df['YearsExperience']), line = 'r')
plt.title('Cube root transformation')
plt.show()
```

```python
stm.qqplot(df['Salary'], line = 'r')
plt.title('No transformation')
stm.qqplot(np.log(df['Salary']), line = 'r')
plt.title('Log transformation')
stm.qqplot(np.sqrt(df['Salary']), line = 'r')
plt.title('Square root transformation')
stm.qqplot(np.cbrt(df['Salary']), line = 'r')
plt.title('Cube root transformation')
plt.show()
```

```python
import statsmodels.formula.api as smf
model = smf.ols('Salary~YearsExperience', data = df).fit()
```

```python
model.summary()
```

```python
#Square Root transformation 
model1 = smf.ols('np.sqrt(Salary)~np.sqrt(YearsExperience)', data = df).fit()
model1.summary()
```

```python
#The R-squared and Adjusted R-squared scores are above 0.85
```

```python
#Cuberoot transformation 
model2 = smf.ols('np.cbrt(Salary)~np.cbrt(YearsExperience)', data = df).fit()
model2.summary()
```

```python
# ---Log transformation
model3 = smf.ols('np.log(Salary)~np.log(YearsExperience)', data = df).fit()
model3.summary()
```

```python
model.params
```

```python
print(model.tvalues,'\n',model.pvalues)
```

```python
model.rsquared,model.rsquared_adj
```

```python
#Residual Analysis----qqplot

import statsmodels.api as sm
sm.qqplot(model.resid, line = 'q')
plt.title('Normal Q-Q plot of residuals of Model without any data transformation')
plt.show()
```

```python
#model follows normal distribution
```

```python
#Comparing different models with respect to the Root Mean Squared Errors
```

```python
from sklearn.metrics import mean_squared_error
model1_pred_y =np.square(model1.predict(df['YearsExperience']))
model2_pred_y =pow(model2.predict(df['YearsExperience']),3)
model3_pred_y =np.exp(model3.predict(df['YearsExperience']))
```

```python
model1_rmse =np.sqrt(mean_squared_error(df['Salary'], model1_pred_y))
model2_rmse =np.sqrt(mean_squared_error(df['Salary'], model2_pred_y))
model3_rmse =np.sqrt(mean_squared_error(df['Salary'], model3_pred_y))
print('model=', np.sqrt(model.mse_resid),'\n' 'model1=', model1_rmse,'\n' 'model2=', model2_rmse,'\n' 'model3=', model3_rmse)
```

```python
rmse = {'model': np.sqrt(model.mse_resid), 'model1': model1_rmse, 'model2': model3_rmse, 'model3' : model3_rmse}
min(rmse, key=rmse.get)
```

```python
# model is Simple Linear regression model that has the minimum RMSE and highest 
#Adjusted R-squared score. Hence, we are going to use model to predict our values
```

```python
# first model results without any transformation
predicted = pd.DataFrame()
predicted['YearsExperience'] = df.YearsExperience
predicted['Salary'] = df.Salary
predicted['Predicted_Salary_Hike'] = pd.DataFrame(model.predict(predicted.YearsExperience))
predicted
```
